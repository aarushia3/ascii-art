```python
from PIL import Image
import numpy as np
import math
```

```python
imagePath = "../images/example1.webp"
im = Image.open(imagePath)
print("Successfully loaded image!")
print("Image Size: ", im.size)
```

```python
imArray = np.asarray(im).astype(int)
imArray = imArray[:480, :640]
print("Successfully loaded image into array!")
print("Pixel array dimensions: ", imArray.shape)
print(imArray)
```

```python
reducedIm = np.zeros_like(imArray)
for i in range(imArray.shape[0]):
    for j in range(imArray.shape[1]):
        reducedIm[i][j] = (imArray[i][j][0] + imArray[i][j][1] + imArray[i][j][2])/3
reducedIm = reducedIm[:, :, 0]
print("Successfully constructed brightness array!")
print("Brightness Array Dimensions: ", reducedIm.shape)
print(reducedIm)
```

```python
charSpectrum = "`^\",:;Il!i~+_-?][}{1)(|\\/tfjrxnuvczXYUJCLQ0OZmwqpdbkhao*#MW&8%B@$"
len(charSpectrum)
```

```python
charArray = np.empty(reducedIm.shape, dtype=str)
for i in range(reducedIm.shape[0]):
    for j in range(reducedIm.shape[1]):
        index = min(int(math.ceil(reducedIm[i][j] / 255 * 65)), 64)
        charArray[i][j] = charSpectrum[index]
print("Successfully constructed ASCII character array!")
print("ASCII Array Dimensions: ", charArray.shape)
print(charArray)
```

```python
for row in charArray:
    print("".join(row))
```
